# **Gender distribution**

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship brtween Parent's Education and studuent's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of 
#the parent have a good impact on thir score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize= (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship brtween Parent's Marital Status and studuent's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible 
#impact on the student score  due to their parent's marital status
```

# **Distribution of Ehnic Groups**

```python
#pandas(pd) loc function use
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA", "groupB", "groupC", "groupD", "groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"],]

print(mlist)
plt.pie(mlist, labels = l, autopct = "%1.2f%%") #(pie means piechar, autopct means percentag value)
plt.title("Distributio of Erhnic Groups")
plt.show()
```
